```python
import numpy as np
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

Eb = 622.25
Fs = 739.99
Ab = 830.61
B = 987.77
Db = 1108.73
Eb = 1244.51
E = 1318.51
```

```python
# Surfaces - Good Day

notes = [Fs*2, E, Eb, B, Db, Db, 0, Fs, Ab, Fs, Ab, B, Ab, 0, Fs, Fs, Ab, Fs, Ab, B, Ab, Fs, Ab, Fs, Eb*0.5, 0, Fs, Ab, Fs, Ab, B, Db, Db, B, B, Ab, B, Db, B, Db, Eb, B, B, E, Eb, Db, B]

durs = [0.4, 0.4, 0.6, 0.6, 0.35, 0.85, 0.4, 0.25, 0.25, 0.25, 0.25, 0.5, 0.75, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.75, 0.3, 0.15, 0.5, 0.25, 0.75, 0.25, 0.25, 0.25, 0.25, 0.25, 0.5, \
        0.35, 0.15, 0.3, 0.3, 0.75, 0.25, 0.25, 0.25, 0.25, 0.4, 0.75, 0.15, 0.25, 0.25, 0.25, 1.0]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
