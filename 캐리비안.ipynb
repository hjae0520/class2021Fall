```python
import numpy as np
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

C = 261.63
D = 293.66
E = 329.63
F = 349.23
G = 392.00

A = 440.00
```

```python
# 캐리비안

notes = [A, C*2, D*2, 0, D*2, 0, D*2, E*2, F*2, 0, F*2, 0, F*2, G*2, E*2, 0, E*2, D*2, C*2, C*2, D*2]

durs = [0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.15]
```

```python
len (durs)
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
